```python
import pandas as pd
from sklearn.metrics import make_scorer
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('glass.csv')
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import RobustScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Type'],axis=1)
y = df['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train['Mg'].shape
```

```python
scaler = RobustScaler()

for col in X_train.columns:
    scaler = scaler.fit(X_train[col].reshape(-1,1))
    X_train[col] = scaler.transform(X_train[col].reshape(-1,1))
    X_test[col] = scaler.transform(X_test[col].reshape(-1,1))
```

```python
X_train
```

```python
from sklearn.model_selection import cross_val_score
params = {'max_depth': None, 'max_features': 0.05, 'min_samples_split': 0.005, 'n_estimators': 200}
clf = RandomForestClassifier(max_depth= None, max_features = 0.05, min_samples_split= 0.005, n_estimators = 200)

clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

print(accuracy_score(y_test,y_pred))
```
